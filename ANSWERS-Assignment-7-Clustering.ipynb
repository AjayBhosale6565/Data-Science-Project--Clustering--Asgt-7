## K-Means Clustering

```python
cr_data.dtypes
```

```python
x = cr_data.values[:,[1,2,3,4]]
```

```python
from sklearn.cluster import KMeans 
from warnings import filterwarnings 
filterwarnings("ignore") 

wsse = [] 
for i  in range(1,11):
    kmeans = KMeans(n_clusters= i, random_state=5)
    kmeans.fit(x) 
    wsse.append(kmeans.inertia_) 
plt.plot(range(1,11),wsse)
plt.title('Elbow Method') 
plt.xlabel('Number of Clusters') 
plt.ylabel('WSSE')
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
cr_data3 = cr_data[['Murder', 'Assault', 'UrbanPop', 'Rape']]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(cr_data3)
```

```python
#DBSCAN clustering
dbscan = DBSCAN(eps=1.2, min_samples=5)
dbscan.fit(scaled_data)

cr_data['DBSCAN_Cluster'] = dbscan.labels_
print(cr_data)
```

#### Inferences: 
1. Overall the crime rate of individial crime is showing to be almost same. 
2. The murder crime happens to be representing highest in proportion.
3. But still highest number of cluster seems to be in case of rape.
